# Divisão dados UNM
## 1. Carregar dados de atribuição UNM

````python
import pandas as pd
````

````python
df_atibuicao_unm = pd.read_table('dados/metadados/frames_selecionados_treinamento_geral.txt', header=None, sep=' ')
df_atibuicao_unm.columns = ['video_file', 'video_frame_txt']
df_atibuicao_unm
````

````python
import re

def extract_video_frame_id(video_frame_txt, extract_video=False, extract_frame=False):
    numbers = re.findall(r'\d+', video_frame_txt)

    if extract_video == False and extract_frame == False:
        raise Exception('Choose video or frame.')

    return_list = []
    if extract_video:
        return_list.append(numbers[0])
    
    if extract_frame:
        return_list.append(numbers[1])

    return return_list
    
df_atibuicao_unm['video_id'] = (
    df_atibuicao_unm
    .video_frame_txt
    .apply(
        lambda txt: extract_video_frame_id(txt, extract_video=True)[0]
    )
    
)

df_atibuicao_unm['frame_id'] = (
    df_atibuicao_unm
    .video_frame_txt
    .apply(
        lambda txt: extract_video_frame_id(txt, extract_frame=True)[0]
    )
)

# extract_video_frame_id(df_atibuicao_unm.video_frame_txt.values[347])
````

````python
df_atibuicao_unm['video_frame_txt_abv'] = 'v' + df_atibuicao_unm['video_id'] + '_f' + df_atibuicao_unm['frame_id']
df_atibuicao_unm['video_frame_txt_abv']
````

````python
df_atibuicao_unm
````

````python
list_without_points = [
    'v063_f1222',
    'v065_f1195',
    'v065_f1239',
    'v065_f1414',
    'v066_f474',
    'v066_f482',
    'v066_f487',
    'v066_f620',
    'v066_f724',
    'v066_f758',
    'v066_f804',
    # 'v068_f1029', # Vini adicionou o ponto
    'v069_f401',
    'v069_f524',
    'v069_f556',
    'v072_f1102',
    'v072_f1105',
    'v072_f1217',
    'v072_f1473',
    'v072_f1524',
    'v072_f1717',
    'v072_f862',
    'v078_f1233',
    # 'v079_f775', # Vini adicionou o ponto
    'v088_f376',
    'v089_f572',
    'v089_f746'
 ]

list_without_masks = [
    'v063_f1222',
    'v065_f1195',
    'v065_f1239',
    'v065_f1414',
    'v066_f474',
    'v066_f482',
    'v066_f487',
    'v066_f620',
    'v066_f724',
    'v066_f758',
    'v066_f804',
    'v069_f401',
    'v069_f524',
    'v069_f556',
    'v072_f1102',
    'v072_f1105',
    'v072_f1217',
    'v072_f1473',
    'v072_f1524',
    'v072_f1717',
    'v072_f862',
    'v078_f1233',
    'v088_f376',
    'v089_f572',
    'v089_f746'
]

frames_without_something = list(set(list_without_points).union(set(list_without_masks)))
df_frames_without_something = pd.DataFrame(frames_without_something, columns=['video_frame_txt'])

df_frames_without_something['has_points'] = df_frames_without_something['video_frame_txt'].apply(lambda txt: txt not in list_without_points)
df_frames_without_something['has_masks'] = df_frames_without_something['video_frame_txt'].apply(lambda txt: txt not in list_without_masks)

df_frames_without_something
````

````python
df_frames_without_something
````

````python
print('Shape antes:',df_atibuicao_unm.shape, df_frames_without_something.shape)
df_atibuicao_unm_merged = df_atibuicao_unm.merge(
    df_frames_without_something,
    left_on='video_frame_txt_abv',
    right_on='video_frame_txt', 
    suffixes=('', '_y'),
    how='left'
)

del df_atibuicao_unm_merged['video_frame_txt_y']

print('Shape antes:', df_atibuicao_unm.shape)
df_atibuicao_unm_merged['has_points'] = df_atibuicao_unm_merged['has_points'].fillna(True)
df_atibuicao_unm_merged['has_masks'] = df_atibuicao_unm_merged['has_masks'].fillna(True)
df_atibuicao_unm_merged
````

````python
def create_label_status(row):
    if row['has_points'] and row['has_masks']:
        return 'both'
    elif row['has_points'] and not row['has_masks']:
        return 'only_points'
    elif not row['has_points'] and row['has_masks']:
        return 'only_masks'
    else:
        return 'none'

df_atibuicao_unm_merged['label_status'] = df_atibuicao_unm_merged.apply(create_label_status, axis=1)
````

## 2. Análise dos dados de atribuição UNM

## 3. Split dos dados em treino, validação e teste

````python
from sklearn.model_selection import train_test_split
````

````python
has_all_points_and_masks = (df_atibuicao_unm_merged['has_points'] == True) & (df_atibuicao_unm_merged['has_masks'] == True)

df_frames_all_labels = df_atibuicao_unm_merged[has_all_points_and_masks].copy()
df_frames_all_labels.reset_index(inplace=True, drop=True)
df_frames_all_labels
````

### Divisão train/teste (80%/20%)

```python

#### Considerando divisão por frame

````python
X_train, X_test = train_test_split(df_frames_all_labels, test_size=0.2, random_state=42)

print('Train shape:', X_train.shape)
print('Test shape:', X_test.shape)
````

#### Considerando divisão por vídeo

````python
video_train, video_test = train_test_split(df_frames_all_labels.video_id.unique(), test_size=0.2, random_state=42)

# print('Video Train:', video_train)
X_train = df_frames_all_labels[df_frames_all_labels.video_id.isin(video_train)] 
X_test = df_frames_all_labels[df_frames_all_labels.video_id.isin(video_test)] 

print('Train shape:', X_train.shape)
print('Test shape:', X_test.shape)
````

````python
cols = ['video_frame_txt_abv','video_id', 'frame_id']
X_train_videos = X_train[cols].reset_index(drop=True)
X_test_videos = X_test[cols].reset_index(drop=True)

X_train_videos.video_id = X_train_videos.video_id.astype(int)
X_test_videos.video_id = X_test_videos.video_id.astype(int)

print('Treino:')
display(X_train_videos)
print('Teste:')
display(X_test_videos)

X_train_videos.to_csv('dados/split/unm_train_test_por_video/train_videos.csv', index=False)
X_test_videos.to_csv('dados/split/unm_train_test_por_video/test_videos.csv', index=False)
````

### K-Fold Cross Validation

#### Considerando a divisão por frame

````python
from sklearn.model_selection import KFold

X_train_val, X_test = train_test_split(df_frames_all_labels, test_size=0.2, random_state=42)

print('Train shape:', X_train_val.shape)
print('Test shape:', X_test.shape)

folds = []
kf = KFold(n_splits=5, shuffle=True, random_state=42)
for train_index, val_index in kf.split(X_train_val):
    fold_data = X_train_val.iloc[val_index]
    folds.append(fold_data)
````

````python
for i, fold in enumerate(folds):
    print(f'Fold {i+1} shape: {fold.shape}')
````

#### Considerando divisão por vídeo

````python
from sklearn.model_selection import KFold

video_id_train_val, video_id_test = train_test_split(df_frames_all_labels.video_id.unique(), test_size=0.2, random_state=42)

X_train = df_frames_all_labels[df_frames_all_labels.video_id.isin(video_id_train_val)] 
X_test = df_frames_all_labels[df_frames_all_labels.video_id.isin(video_id_test)] 

print(f'Train/val videos quantity: {video_id_train_val.shape} | Frames: {X_train.shape}')
print(f'Test video quantity: {video_id_test.shape} | Frames: {X_test.shape}\n')

folds = []
kf = KFold(n_splits=5, shuffle=True, random_state=42)
for i, (train_index, val_index) in enumerate(kf.split(video_id_train_val)):
    val_fold_video_ids = video_id_train_val[val_index]
    val_fold_data = df_frames_all_labels[df_frames_all_labels.video_id.isin(val_fold_video_ids)] 
    print(f'Validation fold {i+1} shape: {val_fold_data.shape} | Video IDs: {val_fold_video_ids} ')
    folds.append(val_fold_data)


print(f'\nTest shape: {X_test.shape} | Video IDs: {video_id_test}')


````
